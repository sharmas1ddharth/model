```
import os
import numpy as np
import mmcv
import torch
from mmcv.runner import load_checkpoint
from mmdet.models import build_detector
from mmdet.apis import inference_detector
from mmdet.core import get_classes
import pycocotools.mask as maskutils
import pytesseract
import camelot as cam
import pandas as pd
from tika import parser
import cv2
from pytesseract import Output
from PIL import Image,ImageEnhance, ImageFilter
import os 
import easyocr
import re
import warnings
warnings.filterwarnings("ignore")

import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import pandas as pd


class ExtractTable():
  def __init__(self, file):
    self.file = file

  def detect_table(self):
    config_path = '/content/CascadeTabNet/Config/cascade_mask_rcnn_hrnetv2p_w32_20e.py'
    model_path = '/content/epoch_36.pth'

    cfg = mmcv.Config.fromfile(config_path)
    cfg.model.pretrained = None
    cfg.data.test.test_mode = True

    # construct the model and load checkpoint
    model = build_detector(cfg.model, test_cfg=cfg.test_cfg)
    model.cfg = cfg
    device = torch.device("cuda:0")
    model = model.to(device)
    model.eval()
    _ = load_checkpoint(model, model_path)
    mask_score_thresh = 0.6

    img = self.file
    img_result = inference_detector(model, img)
    img = mmcv.imread(img)
    result = img_result
    return result

  def get_table(self, x1, x2, y1, y2):
    img = cv2.imread(self.file)
    cropped_image = img[int(x1):int(x2), int(y1):int(y2)]
    return cropped_image

  def show_table(self):
    img = cv2.imread(self.file)
    bbox,_ = self.detect_table()
    bbox_cordinates = bbox[2].tolist()
    for i in range(len(bbox_cordinates)):
      x1=bbox_cordinates[i][1]
      x2=bbox_cordinates[i][3]
      y1=bbox_cordinates[i][0]
      y2=bbox_cordinates[i][2]
      cropped_image = img[int(x1):int(x2), int(y1):int(y2)]
      cv2_imshow(cropped_image)
      print()

  def get_cordinates(self, n):
    bbox,_ = self.detect_table()
    bbox_cordinates = bbox[2].tolist()
    x1=bbox_cordinates[n][1]
    x2=bbox_cordinates[n][3]
    y1=bbox_cordinates[n][0]
    y2=bbox_cordinates[n][2]
    return self.get_table(x1, x2, y1, y2)

  def extract_text(self, n=0):
    self.get_cordinates(n)
    im = Image.open("result.png")
    # custom_config = r'--oem 3 --psm 6'
    custom_config = r'-l eng --oem 3 --psm 6 '
    text = pytesseract.image_to_string(im, config=custom_config)
    return text


  def table(self, n=0):
    img = self.get_cordinates(n)
    custom_config = r'--oem 3 --psm 6'
    ex_text = pytesseract.image_to_string(img, config=custom_config)
    ex_text = ex_text.split('\n')

    data_list = []
    for i in ex_text[0:]:
      c = i.split(' ')
      temp = []
      for i in c:
        if not i.isnumeric() and not i.replace('.', '').isnumeric():
          temp.append(i)
      temp_str = ' '.join(temp)
      final_list = [x for x in c if x not in temp]
      final_list.insert(0, temp_str)
      data_list.append(final_list)

    try:
      try:
        df=pd.DataFrame(data_list,columns=['Item', 'Qty', 'Rate', 'Amount'])
      except:
        df=pd.DataFrame(data_list,columns=['Item', 'Qty', 'Rate','MRP', 'Amount'])
    except:
      try:
        df=pd.DataFrame(data_list,columns=[ex_text[0].split(' ')])
      except:
        try:
          df=pd.DataFrame(data_list,columns=['Item', 'Amount'])
        except:
          df=pd.DataFrame(data_list,columns=['Item','Qty', 'Amount'])
    
    df = df.replace({r'[^\x00-\x7F]+':''}, regex=True)
    return df

  def pdf_table_extract(self): # extract data from pdf
    data = cam.read_pdf(self.file)
    df = pd.DataFrame(data[0].df)
    return df

  def extract_table(self): # main function to extract table
    extension=self.file.split('.')[-1]
    try:
      if extension != "pdf":
        data = self.table()
      else:
        data = self.pdf_table_extract()
      return data
    except:
      return "No tables found"



class ExtractData():
  def __init__(self, file):
    self.file = file



  def img_table_extract(self): # convert image to pdf then extract
    pdf = pytesseract.image_to_pdf_or_hocr(f"{self.file}", extension='pdf')
    with open(f'{self.file}.pdf', 'w+b') as f:
        f.write(pdf) # pdf type is bytes by default
  
    path = os.path.join(parent_dir, directory)
    data = cam.read_pdf(f"{self.file}.pdf")
    df = pd.DataFrame(data[0].df)
    df = df.replace({r'[^\x00-\x7F]+':''}, regex=True)
    df1 = df[df.isnull().sum(axis=1) < 2]
    return df1

  
  def extract_pdf_data(self): # extract data
    raw = parser.from_file(self.file)
    text=raw['content']
    return text

  def extract_img_data(self):
    im = Image.open(self.file)
    text = pytesseract.image_to_string(im)
    return text
    

  def extract_table(self): # main function to extract table
    extension=self.file.split('.')[-1]
    try:
      if extension != "pdf":
        data = self.img_table_extract()
      else:
        data = self.pdf_table_extract()
      return data
    except:
      return "No tables found"


  def extract_data(self): # main function to extract data
    extension=self.file.split('.')[-1]
    if extension != "pdf":
      # self.im_ehn()
      data = self.extract_img_data()
    else:
      data = self.extract_pdf_data()
    return data


  def get_data(self):
    text = self.extract_data()
    char_to_replace = ['~', '!', '#', '^', '*', '_', ';']
    for c in char_to_replace:
      text = text.lower().strip().replace(c, ' ')
      
    text=text.lower().replace(':\n', ':').replace('\n\n', '\n')

    store_name = re.compile(r'sold.+\s\S.+|fro.+\s\S.+inc|f.*m.*:.*|.*market|li.st.+|.*mart|.*cafe|.*restaurant|[a-z]+.&.[a-z]+').findall(text)
    store_address = re.compile(r'sold.+.\s\S*.*\s\S.*\s.*\s\S.*\d{6}|.+,.+\d{6}|.+.\s\S*.*\s\S.*\s.*\s\S.*\d{6}|.+,.+\d{5}').findall(text)
    bill_number = re.compile(r'invoice\s?n.+:.+|bill\s?n.+:.+|invoice no-\d+|no.+\d{5}||no.\s\d{}').findall(text)
    bill_date = re.compile(r'o[a-z].*date.+\d+\.\d+\.\d+|i[a-z].*date.+\d+\.\d+\.\d+|o[a-z].*date.+\d+-\d+-\d+|\d{2}-\d{2}-\d{4}|\d{2}/\d{2}/\d{4}|\d{2}-[a-z].+-\d{2}|\d{2}-\d{2}-\d{2}|\d{2}/\d{1}/\d{4}|[a-z]+.\d{2},\d{4}|[a-z]+-\d{2}-\d{4}').findall(text)
    total = re.compile(r'total.+\s\S\d+.*|total.\d.+|\d+\.\d+').findall(text) #total

    store_name_c = self.store_name_clean(store_name)
    store_address_c = self.store_address_clean(store_address)
    bill_number_c  = self.invoice_clean(bill_number)
    bill_date_c = self.date_clean(bill_date)
    total_c = self.basic_clean(total)

    ls = {'store_name':store_name_c, 
            'store_address':store_address_c, 
            'bill_number':bill_number_c, 
            'bill_date':bill_date_c, 
            'total':total_c}
      
    return ls

  def basic_clean(self, lst):
    if len(lst) >= 1:
      return lst[0].replace('\n', ', ').replace(',,', ', ')
    else:
      return lst
    
  def store_name_clean(self, lst):
    if len(lst) >= 1:
      return lst[0].replace('sold by:, ', '').replace('sold by:\n', '')
    else:
      return lst

  def store_address_clean(self, lst):
    if len(lst) >= 1:
      return lst[0].replace('  ,  , ', '').replace('\n', '').replace('   ', '').replace('sold by:', '')
    else:
      return lst

  def invoice_clean(self, lst):
    # invoice_data = self.basic_clean(lst)
    if len(lst) >= 1:
      return lst[0].replace('invoice number : ', '').replace('invoice number:', '').replace('no: ', '').replace('no.', '').replace('invoice no, : ', '').replace('invoice  : ', '').replace('invoice no-','').upper()
    else:
      return lst
  
  def total_clean(self, lst):
    if len(lst) == 1:
      total = lst[0].replace('total:', '')
      return total
    elif len(lst) > 1:
      totals = [float(amount) for amount in lst]
      return max(totals)
    else:
      return lst

  def date_clean(self, lst):
    if len(lst) >= 1:
      return lst[0].replace('order date: ', '').upper()
    else:
      return lst
    


Invoice = type('Invoice', (ExtractTable,ExtractData), dict(Invoice='Invoice'))
```
